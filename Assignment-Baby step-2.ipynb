```python
import pandas as pd
```

```python
# call the function read_csv() and send the parameter Salary_Data.csv
# The values are assigned to the variable dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# view the values in dataset
dataset
```

```python
# Here split the input & output variable 
# YearsExperience is the input values to assign the variable independent
independent=dataset[["YearsExperience"]]
```

```python
# view the input values
independent
```

```python
# Assign the output values to  variable dependent
dependent=dataset[["Salary"]]
```

```python
#View the output values
dependent
```

```python
# split the train and test dataset
# Here the library sklearn support the machine learning
from sklearn.model_selection import train_test_split
# x_train,x_test is input variable for training dataset and y_train,y_test is output variable for test dataset
x_train,x_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30, random_state=0)
```

```python
# Model creation
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
# Learning Phase
regressor.fit(x_train,y_train)# here only weight and bias calculated
```

```python
# check the weight and bias values
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
#Evalution Metrics
y_pred=regressor.predict(x_test)
y_pred
```

```python
# Test the model if it is working perfectly
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
# Save the model
import pickle
filename="Finalized_model_Linear.sav"
```

```python
pickle.dump(regressor, open (filename,'wb'))
```

```python
#Phase 2
#Load the Model
loaded_model= pickle.load (open("Finalized_model_Linear.sav",'rb'))
experience=int(input("Enter the Years of Experience:"))
result=loaded_model.predict([[experience]])
print ("The salary of ",experience,"years experience is",result)
```
