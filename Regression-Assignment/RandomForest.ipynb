```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset['sex'].replace(['female','male'],[1,0],inplace=True)
dataset['smoker'].replace(['yes','no'],[1,0],inplace=True)
dataset
```

```python
independent=dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,criterion='absolute_error',max_features='sqrt',bootstrap=True, random_state=0)
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename="Finalized_Model_RFR.sav"
pickle.dump(regressor, open (filename,'wb'))
```

```python
loaded_model= pickle.load(open("Finalized_model_RFR.sav",'rb'))
age=int(input("Enter your Age:"))
sex_male=int(input("Enter gender Male(0) or female(1) enter 0 or 1:"))
bmi=float(input("Enter your BMI:"))
children=int(input("Enter the No of children:"))
smoker_yes=int(input ( "Smoker Yes(1) or No(0) Enter 0 or 1:"))
result=loaded_model.predict([[age,sex_male,bmi,children,smoker_yes]])
result                
```
