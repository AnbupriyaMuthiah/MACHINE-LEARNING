```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset['sex'].replace(['female','male'],[1,0],inplace=True)
dataset['smoker'].replace(['yes','no'],[1,0],inplace=True)
dataset
```

```python
independent=dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error', splitter='best',max_features="log2")
regressor=regressor.fit(x_train,y_train)
```

```python
#import matplotlib.pyplot as plt 
#from sklearn import tree
#tree.plot_tree(regressor)
#plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
